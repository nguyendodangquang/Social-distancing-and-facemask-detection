```python
%cd /content/gdrive/MyDrive/yolov4_tiny/darknet

import glob2
import math  
import os
import numpy as np
np.random.seed(42)
files_train = []
files_val = []
for ext in ["*.png", "*.jpeg", "*.jpg"]:
    image_files_train = glob2.glob(os.path.join("data/train/", ext))
    files_train += image_files_train
np.random.shuffle(files_train)
with open("train.txt", "w") as f:
    for idx in np.arange(len(files_train)):
        if (os.path.exists(files_train[idx][:-3] + "txt")):
            f.write(files_train[idx]+'\n')

for ext in ["*.png", "*.jpeg", "*.jpg"]:
    image_files_val = glob2.glob(os.path.join("data/valid/", ext))
    files_val += image_files_val
np.random.shuffle(files_val)
with open("val.txt", "w") as f:
    for idx in np.arange(len(files_val)):
        if (os.path.exists(files_val[idx][:-3] + "txt")):
            f.write(files_val[idx]+'\n')

# with open("train.txt", "w") as f:
#     for i in train:
#         f.write(files[i]+'\n')
# # with open("train.txt", "w") as f:
# #     f.write(files[idx]+'\n')

# # Tạo file val.txt
# with open("val.txt", "w") as f:
#   for idx in rand_idx:
#     if (os.path.exists(files[idx][:-3] + "txt")):
#       f.write(files[idx]+'\n')
```
